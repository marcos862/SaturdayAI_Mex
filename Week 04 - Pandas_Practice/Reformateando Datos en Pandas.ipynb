### Como se dice que solo se usaron 3 columnas del dataset original, vamos a remover todas las demas columnas y quedarnos con las que son de interes.

### La columna **Material**, la vamos a renombrar como **CÓDIGO**

Esta es otra opción (b) que usando apply y lambda, extrae del string el año y el mes y crea directamente las columnas de Año y Mes

```python
#
# Esta opción es mucho mas rápida que la opción (a), el problema es que se hacen muchas operaciones en la misma linea
# y no se podría entender exactamente lo que hace
#
df_int_col['Año'] = df_int_col['Date'].apply(lambda x: int(x.split('/')[2][0:4]))
df_int_col['Mes'] = df_int_col['Date'].apply(lambda x : calendar.month_abbr[int(x.split('/')[0])].upper())
df_int_col.head()
```

### Ahora, ya que tenemos casi el dataframe como requerimos, vamos a eliminar la columna de Date y hacer un pivot Table tomando el mes como el pivote

Eliminamos primero la columna **Date**

```python
df_ym = df_int_col.drop(['Date'], axis=1)
df_ym.head()
```

Pivoteamos la columna de **Month**, hay que tener cuidado de cambiar el parámetro aggfunc, el cual, por default, calcula los promedios, y lo que nosotros queremos es la suma

```python
df_final = df_ym.pivot_table(values = 'Units', index = ['CÓDIGO', 'Año'], columns = 'Mes', aggfunc = np.sum, fill_value=0)
df_final.head()
```

### Ahora, vamos a Acomodar los nombres de los meses para que esten al mismo "nivel" que **CÓDIGO**

```python
df_final = df_final.reset_index()
df_final.head()
```

Por último, vamos a eliminar la columna de Month, que ya no hace sentido

```python
df_final.columns.name = None
df_final.head()
```

### Vamos a comprobar que el **"Quality Report"** de nuestra tabla generada, es igual a la que nos proporcionaron

Para eso, creamos la función de **Quality Report**

```python
def quality_report(data):

    """This method will do a basic data quality report for a data frame"""
        
    if (type(data) != pd.core.frame.DataFrame):
        raise TypeError("Data must be pandas.core.frame.DataFrame")
    else: 
        columns = list(data.columns.values)
        data_type = pd.DataFrame(data.dtypes, columns=['Data type'])
        missing_data = pd.DataFrame(
        data.isnull().sum(), columns=['missing values'])
        present_data = pd.DataFrame(data.count(), columns=['present values'])
        unique_values = pd.DataFrame(columns=['unique values'])
        minimum_values = pd.DataFrame(columns=['minimum values'])
        max_values = pd.DataFrame(columns=['maximun values'])
        
        for i in columns:
            unique_values.loc[i] = [data[i].nunique()]
            try:
                minimum_values.loc[i] = [data[i].min()]
                max_values.loc[i] = [data[i].max()]
            except:
                pass
        
        DQ_report = data_type.join(missing_data).join(present_data).join(
        unique_values).join(minimum_values).join(max_values)
    
    return DQ_report
```

Corremos el Quality Report en nuestro DataFrame

```python
quality_report(df_final)
```
